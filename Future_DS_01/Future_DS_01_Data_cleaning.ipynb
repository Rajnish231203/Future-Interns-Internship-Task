```python

import re

# Step 1: Define a function to remove words containing any numerical data
def remove_numbers(text):
    if pd.isnull(text):
        return text
    # Remove any word that contains at least one digit
    return ' '.join([word for word in text.split() if not re.search(r'\d', word)])

# Step 2: Apply this function to Description column
Data['Cleaned_Description'] = Data['Description'].apply(remove_numbers)

# Step 3: Check results
Data[['Description', 'Cleaned_Description']].head(10)
```

```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

# Ensure required resources are available
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# Predefined list of common adjectives
COMMON_ADJECTIVES = {
    'small', 'large', 'big', 'tiny', 'huge', 'mini', 'red', 'blue', 'green', 'yellow',
    'black', 'white', 'pink', 'brown', 'grey', 'gray', 'beautiful', 'pretty', 'cute',
    'nice', 'sweet', 'lovely', 'modern', 'vintage', 'classic', 'new', 'old', 'soft',
    'hard', 'light', 'heavy', 'round', 'square', 'long', 'short', 'wide', 'narrow',
    'good', 'bad', 'amazing', 'wonderful', 'fantastic', 'cheap', 'expensive',
    'wooden', 'ceramic', 'metal', 'glass', 'plastic', 'shiny', 'dull', 'bright', 'dark'
}

# Remove adjectives using POS tagging
def remove_adjectives_nltk(text):
    try:
        tokens = word_tokenize(text)
        pos_tags = pos_tag(tokens)
        return ' '.join([word for word, tag in pos_tags if tag not in ['JJ', 'JJR', 'JJS']])
    except:
        return text

# Remove adjectives from predefined list
def remove_adjectives_list(text):
    words = re.findall(r'\b\w+\b', text.lower())
    return ' '.join([word for word in words if word not in COMMON_ADJECTIVES])

# Hybrid method combining both
def remove_adjectives_hybrid(text):
    text = str(text)
    no_adj_nltk = remove_adjectives_nltk(text)
    return remove_adjectives_list(no_adj_nltk).upper()

# Apply hybrid method to your DataFrame
Data['Cleaned_Description_No_Adjectives'] = Data['Cleaned_Description'].apply(remove_adjectives_hybrid)
```

```python
import pandas as pd

# Category keywords definition
CATEGORY_KEYWORDS = {
    'kitchen_dining': [
        'mug', 'cup', 'glass', 'bottle', 'jar', 'container', 'food', 'bread', 'bin',
        'bowl', 'plate', 'dish', 'serving', 'cake', 'stand', 'tier',
        'spoon', 'spoons', 'measuring', 'fork', 'knife', 'cutlery', 'set',
        'kitchen', 'cook', 'cooking', 'baking', 'chef',
        'coffee', 'tea', 'wine', 'drink', 'beverage', 'teapot', 'kettle',
        'herb', 'mint', 'diner', 'lunch', 'dinner', 'breakfast',
        'goldfish', 'pet', 'feeding'
    ],

    'edible_food_snacks': [
        'chocolate', 'candy', 'sweet', 'sweets', 'snack', 'snacks',
        'biscuit', 'cookie', 'cake', 'pastry', 'muffin', 'treat',
        'dessert', 'lollipop', 'marshmallow', 'toffee', 'sugar',
        'gummy', 'chew', 'cracker', 'popcorn', 'donut',
        'food', 'eat', 'edible', 'yum', 'yummy', 'delicious',
        'fruit', 'berry', 'strawberry', 'flavor', 'flavour'
    ],

    'Home Decor': [
        't-light', 'tlight', 'candle', 'holder', 'light', 'lighting', 'lamp',
        'hanging', 'wick', 'boxed', 'beads', 'morris', 'botanical',
        'decoration', 'decorative', 'decor', 'ornament', 'display',
        'heart', 'star', 'natural', 'vintage', 'traditional',
        'sign', 'metal', 'slate', 'tile', 'hanging', 'wall', 'board',
        'memo', 'retrospot', 'design'
    ],
    
    'textile_fabric_bags': [
        'fabric', 'textile', 'cotton', 'silk', 'polyester', 'wool', 'lace',
        'ribbon', 'ribbons', 'thread', 'yarn', 'knitting', 'nancy', 'crochet',
        'bag', 'handbag', 'shopper', 'jumbo', 'lunch', 'gift', 'carrier',
        'purse', 'pouch', 'tote', 'shopping',
        'doilies', 'doiley', 'paisley', 'flower', 'cushion', 'pillow',
        'curtain', 'blind', 'mat', 'doormat', 'door', 'welcome'
    ],
    
    'storage_organization': [
        'storage', 'container', 'tin', 'tins', 'box', 'boxes', 'case', 'cases',
        'organizer', 'basket', 'drawer', 'knob', 'shelf', 'rack', 'holder',
        'trinket', 'pot', 'keep', 'store', 'organize', 'filing'
    ],
    
    'garden_outdoor': [
        'garden', 'gardening', 'spade', 'tool', 'outdoor', 'plant', 'flower',
        'cottage', 'herb', 'marker', 'mint', 'botanical', 'seed',
        'picnic', 'beach', 'outdoor', 'skittles', 'wooden', 'set',
        'birdhouse', 'bird', 'house', 'tree', 'nature', 'green'
    ],
    
    'craft_stationery': [
        'craft', 'feltcraft', 'stencil', 'alphabet', 'stamp', 'set',
        'stickers', 'patches', 'iron', 'paper', 'collage', 'cups',
        'notebook', 'book', 'memo', 'board', 'pen', 'pencil', 'ruler',
        'calculator', 'office', 'school', 'educational', 'learning'
    ],
    
    'toys_games': [
        'toy', 'doll', 'princess', 'lola', 'charlie', 'figures',
        'rabbit', 'painted', 'teddy', 'bear', 'stuffed',
        'game', 'bingo', 'set', 'puzzle', 'play', 'fun', 'entertainment',
        'block', 'building', 'word', 'educational'
    ],
    
    'seasonal_holiday': [
        'christmas', 'xmas', 'holiday', 'festive', 'traditional', 'winter',
        'santa', 'reindeer', 'snow', 'tree',
        'easter', 'bunny', 'eggs', 'spring', 'chick',
        'valentine', 'halloween', 'thanksgiving', 'celebration'
    ],
    
    'jewelry_accessories': [
        'jewelry', 'jewellery', 'beaded', 'crystal', 'necklace', 'bracelet',
        'ring', 'charm', 'pendant', 'accessory', 'earring', 'brooch',
        'watch', 'keyring', 'key', 'chain', 'fashion'
    ],
    
    'bathroom_personal_care': [
        'bathroom', 'washroom', 'toilet', 'bath', 'shower', 'soap',
        'towel', 'mirror', 'cabinet', 'hygiene',
        'personal', 'care', 'beauty', 'cosmetic', 'health', 'wellness',
        'water', 'bottle', 'hot', 'comfort', 'scottie', 'dog'
    ],
    
    'office_business': [
        'office', 'business', 'desk', 'chair', 'table', 'filing',
        'document', 'folder', 'binder', 'clip', 'stapler',
        'sign', 'metal', 'business', 'professional', 'commercial',
        'cash', 'carry', 'rex', 'postage', 'carriage'
    ],
    
    'books_media': [
        'book', 'books', 'reading', 'story', 'novel', 'magazine',
        'newspaper', 'journal', 'diary', 'notebook',
        'cd', 'dvd', 'music', 'film', 'movie', 'media', 'entertainment'
    ],
    
    'art_creativity': [
        'art', 'paint', 'brush', 'canvas', 'drawing', 'sketch',
        'color', 'colour', 'creative', 'artistic',
        'frame', 'picture', 'photo', 'image', 'poster', 'print',
        'blackboard', 'black', 'board', 'white', 'finish'
    ],
    
    'health_wellness': [
        'health', 'wellness', 'medical', 'first', 'aid', 'medicine',
        'vitamin', 'supplement', 'therapy', 'exercise',
        'comfort', 'relax', 'stress', 'relief', 'massage', 'spa'
    ]
}

def assign_category(row):
    """Assign category to a product based on description and word columns"""
    
    # Get description text
    description = ''
    for col_name in ['Cleaned_Description']:
        if col_name in row and pd.notna(row[col_name]):
            description = str(row[col_name]).lower()
            break
    
    # Get individual words from Word_1 to Word_8 columns
    individual_words = []
    for i in range(1, 9):
        word_col = f'Word_{i}'
        if word_col in row and pd.notna(row[word_col]) and str(row[word_col]) != 'None':
            individual_words.append(str(row[word_col]).lower())
    
    # Score each category
    category_scores = {}
    
    for category, keywords in CATEGORY_KEYWORDS.items():
        score = 0
        
        for keyword in keywords:
            # Check description for exact word matches
            if f' {keyword} ' in f' {description} ':
                score += 3
            # Check description for partial matches
            elif keyword in description:
                score += 2
            
            # Check individual word columns
            for word in individual_words:
                if keyword == word:
                    score += 2
                elif keyword in word and len(keyword) > 3:
                    score += 1
        
        category_scores[category] = score
    
    # Return category with highest score, or 'others' if no matches
    if category_scores and max(category_scores.values()) > 0:
        return max(category_scores, key=category_scores.get)
    else:
        return 'others'
```
